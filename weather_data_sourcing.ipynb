## Load historical weather data for a given time period

```python
def get_stations_nearby(lat, lon):

    """
    Get the nearby weather stations for the given latitude and longitude

    Parameters:
    ----------  
    lat : float
        Latitude of the location
    lon : float
        Longitude of the location

    Returns:
    -------
    stations : list
        List of nearby weather stations

    """
    # Get nearby weather stations
    stations = Stations()

    # Get the German weather stations
    stations = stations.region('DE')

    # Get the nearby stations for the given latitide and longitude
    stations = stations.nearby(lat,lon)

    # Fetch 1 station
    count = stations.region('DE').count()
    stations = stations.fetch(5, sample=True)

    return stations, count
```
